```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Define the dataset
data = pd.read_excel("E:/AKTUARIA FINAL SEASON/DATA/fpdata.xlsx", sheet_name="x1_new")
data = np.array(data)

# Split the data into features and target variable
X = data[:, :-1]
y = data[:, -1]
```

```python
def modelling(X_train, y_train, X_test, y_test, epochs, loss_function, metrics, optimizer, activation, hidden_node):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Input(shape=(X_train.shape[1],)),
        tf.keras.layers.Dense(hidden_node, activation=activation),
        tf.keras.layers.Dense(1)
    ])
    
    model.compile(loss=loss_function, optimizer=optimizer, metrics=[metrics])
    
    history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test), batch_size=1)
    
    # Save model
    model.save(f"model_{hidden_node}.h5")
    
    final_loss = history.history['loss'][-1]
    weights = model.get_weights()
    
    return final_loss, weights[0], weights[1], weights[2], weights[3]

# Assuming X_train, y_train, X_test, y_test are defined

model_sum = pd.DataFrame(columns=['mse', 'hl_weights', 'hl_bias', 'ol_weights', 'ol_bias'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Adjusted code to use specific train and test data
X_train_specific = X_scaled[:238]
y_train_specific = y[:238]
X_test_specific = X_scaled[238:442]
y_test_specific = y[238:442]

for n in range(1, 21):
    a, b, c, d, e = modelling(X_train=X_train_specific,
                               y_train=y_train_specific,
                               X_test=X_test_specific,
                               y_test=y_test_specific,
                               epochs=100,
                               loss_function='mse',
                               metrics='mae',
                               optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.01),
                               activation='relu',
                               hidden_node=n)
    
    model_sum.loc[n] = [a, b, c, d, e]
```
